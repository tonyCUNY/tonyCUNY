### **The FED Funds Rate (FRED) (Federal Reserve Board)**

The Fed Funds Rate is obtained from [Federal Reserve Economic Database](https://fred.stlouisfed.org/series/FEDFUNDS) using the pandas_datareader package.

The federal funds rate represents the range of target interest rates established by the Federal Open Market Committee (FOMC). This target rate determines the rate at which commercial banks borrow and lend their excess reserves to each other overnight.

During a recession, the Federal Reserve may lower the federal funds rate to encourage economic growth. From the plot below, we can see that the fed rate decreased during the event of a recession.


```python
import matplotlib.pyplot as plt
from matplotlib import style
from pandas_datareader.data import DataReader
from datetime import date

style.use('ggplot')

series_code_ffr = 'FEDFUNDS'  # Federal Funds Effective Rate
series_code_recession = 'USRECP'  # Recession indicator
data_source = 'fred'
start_date = date(1970, 1, 1)

# Fetching data for Federal Funds Effective Rate
df_ffr = DataReader(series_code_ffr, data_source, start_date)

# Fetching data for Recession indicator
df_recession = DataReader(series_code_recession, data_source, start_date)

# Plotting Federal Funds Effective Rate
df_ffr.plot()
plt.xlabel(' ')
plt.ylabel('Rate')
plt.title('Federal Funds Effective Rate with recession times shaded')

# Shading recession periods
max_y = df_ffr.max()
recession_values = df_recession['USRECP'].values
plt.fill_between(df_recession.index, 0, max_y, where=recession_values == 1, facecolor='blue', alpha=0.2)

plt.show()
```

### **The Consumer Price Index (CPI) (Bureau of Labor Statistics)**

The Consumer Price Index (CPI) data is retrived from [Federal Reserve Economic Database](https://fred.stlouisfed.org/series/CPIAUCSL) using the pandas_datareader package.

The CPI measures the average change over time in the prices paid by urban consumers for a market basket of consumer goods and services.

One of the Fed's mandate focuses on maintaining stable prices. The Federal Reserve aims to keep inflation in check, ensuring that prices remain relatively constant over time.

Fromt the following plot, we can see the CPI increase steadily overtime.

```python
# Get the data from FRED
series_code_cpi = 'CPIAUCSL'  # Consumer Price Index for All Urban Consumers

# Fetching data for Consumer Price Index
df_cpi = DataReader(series_code_cpi, data_source, start_date)

# Plotting Consumer Price Index
df_cpi.plot()
plt.xlabel(' ')
plt.ylabel('Index')
plt.title('Consumer Price Index for All Urban Consumers: All Items in U.S. City Average')

# Shading recession periods
max_y = df_cpi.max()
recession_values = df_recession['USRECP'].values
plt.fill_between(df_recession.index, 0, max_y, where=recession_values == 1, facecolor='blue', alpha=0.2)

plt.show()
```

## **Unemployment Rate (Bureau of Labor Statistics)**

Another mandate of the Federal Reserve is to maximize employment. The objective is to establish an economic environment that fosters sustainable employment at its highest level. This involves ensuring that as many individuals as possible, who are seeking employment, can find suitable jobs. The ultimate goal is to maintain a reasonable unemployment rate, thereby contributing to overall economic well-being and stability.

We can see although there are spikes in certain years, the unemployment rate would always decrease to 5% or below.

The Labor Force Participation Rate estimates the proportion of people who are either actively seeking employment or are currently employed. During the latter half of the 20th century, it exhibited a consistent upward trend. However, when the Great Recession struck in 2008, the participation rate experienced a prolonged decline, eventually stabilizing at approximately 63% by 2013



```python
# Get the data from FRED
series_code_unem = ['UNRATE', 'CIVPART']  # 'Unemployment Rate', 'Participation Rate'

# Fetching data for 'Unemployment Rate', 'Participation Rate'
df_econ_data = DataReader(series_code_unem, data_source, start_date)

# Assign new column labels
df_econ_data.columns = ['Unemployment Rate', 'Participation Rate']

# Plotting 'Unemployment Rate', 'Participation Rate'
df_econ_data.plot(subplots=True, title='Labor Market', ylabel='Index')
plt.xlabel(' ')
plt.show()
```
